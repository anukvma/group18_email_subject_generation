```python
import transformers
from datasets import load_dataset, load_metric, Dataset,DatasetDict
```

```python
#Microsoft/phi-2 is a Small lang. model(SLM) with 2.7 Billion parameters ; Hosted on - Hugging Face model hub
sModel_name = "microsoft/phi-1"
bTrust_remote_code = True
```

```python
#Load the tokenizer
#trust_remote_code - Allows execution of code from the tokenizer files
tokenizer = AutoTokenizer.from_pretrained(sModel_name, trust_remote_code=bTrust_remote_code)
#Load the pre-trained model
model = AutoModelForCausalLM.from_pretrained(sModel_name, trust_remote_code=bTrust_remote_code)
```

```python
# To display summary
!pip install torchinfo
from torchinfo import summary
summary(model)
```

```python
#Example test
sInput_Text_Prompt = "Once upon a time" #"church goers"
sReturn_Tensors_Format = "pt" #PyTorch tensors
#To convert input text into MLModel usable format (i.e., pytorch tensors)
inputs = tokenizer(sInput_Text_Prompt, return_tensors=sReturn_Tensors_Format)

#Convert the model's output (token ID format) back into readable text
nMax_No_New_Tokens = 50
#Generate sequence of tokens
outputs = model.generate(**inputs, max_new_tokens=nMax_No_New_Tokens)

bSkip_Special_Tokens = True
#First sequence of tokens (outputs[0]) generated, without special tokens like padding, etc
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=bSkip_Special_Tokens)

model.name_or_path
```
